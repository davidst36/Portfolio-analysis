```python
import pandas as pd
import cufflinks as cf
import yfinance as yf
tlt = yf.download("TLT", start="2024-04-01", end=pd.to_datetime("today"))
tlt.to_csv("tlt.csv")
tlt_data = pd.read_csv("tlt.csv", index_col=[0], parse_dates=[0])
```

```python
import warnings
warnings.filterwarnings("ignore", message="DatetimeIndex.format is deprecated", category=FutureWarning)

qf = cf.QuantFig(df = tlt_data)
qf.add_sma(periods = 20)
qf.add_bollinger_bands(periods = 20, boll_std = 2)
qf.iplot()
```
